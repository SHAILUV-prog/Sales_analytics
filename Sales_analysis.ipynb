```python
import numpy as np 
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding="latin1") ## Loading the dataset
```

```python
df.head()  ## This will display the first five rows of the dataframe
```

```python
df.describe()  ## This will give statistical summary of numerical columns
```

```python
df.isnull().sum()  ## This will check for any null values
```

```python
df.drop_duplicates(inplace=True)  ## Dropping duplicate rows if any
```

```python
df.isnull() ## This will check for any remaining null values
```

```python
df["ORDERDATE"]=pd.to_datetime(df["ORDERDATE"])
df["Year"]=df["ORDERDATE"].dt.year
df.sort_values("ORDERDATE",inplace=True)
print(df)  ## This will sort the dataframe by ORDERDATE in ascending order
```

```python
df["MONTH"]=df["ORDERDATE"].dt.to_period("M")
print(df)  ## This will add a new column 'MONTH' with year and month information
```

```python
total_sales=df["SALES"].sum()
print(total_sales)  ## This will give total sales amount
```

```python
df.shape[0]  ## This will give number of rows in the dataframe
```

```python
rows, columns = df.shape
print(rows, columns)   ## This will give number of rows and columns in the dataframe
```

```python
avg_sales=df["SALES"].mean()
print(avg_sales)  ## This will give average sales amount
```

```python
monthly_sales =df.groupby("MONTH")["SALES"].sum()
print(monthly_sales)  ## This will give total sales for each month
```

```python
plt.figure(figsize=(10,6))  ## Setting the figure size
```

```python
monthly_sales.index = monthly_sales.index.to_timestamp()
```

```python
plt.title("Monthly Sales Over Time")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.plot(monthly_sales) ## This will plot the monthly sales over time
plt.xticks(rotation=45)  ## Rotating x-axis labels for better readability
plt.tight_layout() ## Adjusting layout to prevent clipping of labels    
plt.grid()  ## Adding grid for better visualization
plt.show()  ## This will plot the monthly sales over time
```

```python
top_product=df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False) ## This will give total sales for each product line sorted in descending order
print(top_product)
```

```python
plt.figure(figsize=(10,5))  ## Setting the figure size
sns.barplot(x=top_product.index, y=top_product.values)  ## This will create a bar plot for product lines vs their total sales
plt.title("Sales vs revenue")  ## Setting the title of the plot
plt.xlabel("Product name")  ## Setting the x-axis label
plt.ylabel("Product value")  ## Setting the y-axis label
plt.xticks(rotation=45)  ## Rotating x-axis labels for better readability
plt.show()
```

```python
country_sales=df.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False) ## This will give total sales for each country sorted in descending order 
print(country_sales)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=country_sales.index,y=country_sales.values)
plt.title("Sales by country")
plt.xlabel("Country")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

ts = df.groupby(pd.Grouper(key="ORDERDATE", freq="M"))["SALES"].sum()

model = ExponentialSmoothing(ts, trend="add", seasonal="add", seasonal_periods=12)
fit = model.fit()
forecast = fit.forecast(12)
```

```python
plt.figure(figsize=(12,6))
plt.plot(ts, label="Actual Sales") ## Plotting actual sales
plt.plot(forecast, label="Forecast", linestyle="--")  ## Plotting forecasted sales
plt.title("Next 12 Month Sales Forecast")  ## Setting the title of the plot
plt.legend()  ## Adding legend to the plot
plt.show()
```
